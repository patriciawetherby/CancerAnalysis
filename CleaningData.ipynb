```python
import numpy as np
import pandas as pd
```

```python
incidents = pd.read_csv('incd.csv')
income = pd.read_csv('income.csv')
deaths = pd.read_csv('death.csv')
```

```python
incidents.drop(index=0,axis=0) # Remove Total Row
# Remove unwanted values
a = incidents[incidents['Age-Adjusted Incidence Rate(Ê) - cases per 100,000'] == '*'].index
b = incidents[incidents['Age-Adjusted Incidence Rate(Ê) - cases per 100,000'] == '* '].index
c = incidents[incidents['Age-Adjusted Incidence Rate(Ê) - cases per 100,000'] == ' '].index

incidents.drop(a,inplace=True)
incidents.drop(b,inplace=True)
incidents.drop(c,inplace=True)

incidents = incidents[incidents['Average Annual Count'].str.contains('3 or fewer')==False]
incidents = incidents[incidents['Age-Adjusted Incidence Rate(Ê) - cases per 100,000'].str.contains('#')==False]
incidents = incidents[incidents['Age-Adjusted Incidence Rate(Ê) - cases per 100,000'].str.contains('_')==False]
```

```python
income = income.rename(columns=({'County':'County_Name','Median':'Median_Income'})) # Rename income columns to match
```

```python
# Initialize columns
cols_concat = ['County_Name', 'State_Name']

# Convert them to type str
income[cols_concat] = income[cols_concat].astype('str')

# Then concatenate them as follows
income['County'] = income[cols_concat].T.agg(', '.join)
```
